```python
# outliers cleaning
describe = df.describe()

def cleanColumn(data,columns, thr = 2):
    column_desc = describe[columns]

    q3 = column_desc[6]
    q1 = column_desc[4]
    IQR = q3 - q1

    top_limit_clm = q3 + thr * IQR
    bottom_limit_clm = q1 - thr * IQR

    filter_clm_bottom = bottom_limit_clm < data[columns]
    filter_clm_top = data[columns] < top_limit_clm
    filters = filter_clm_bottom & filter_clm_top

    data = data[filters]

    print("{} of dataset after column {}".format(data.shape, columns))

    return data

for i in df.columns:
    data = cleanColumn(df, i)
```

Oversampling

```python
from imblearn.over_sampling import SMOTE

Features = df.drop('Outcome', axis=1)
Target = df['Outcome']


smote = SMOTE(random_state=42)

# Apply SMOTE to generate synthetic samples
Features_oversampled, Target_oversampled = smote.fit_resample(Features, Target)


df1 = pd.concat([pd.DataFrame(Features_oversampled), pd.DataFrame(Target_oversampled, columns=['Outcome'])], axis=1)


print(df1['Outcome'].value_counts())
```

```python
x = df1.drop(columns = 'Outcome', axis=1)
y= df1['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, stratify=y, random_state=0)
```

LogisticRegression



```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

lr = LogisticRegression()

param_grid = {
    'C': [0.001, 0.01, 0.05, 0.1, 0.5, 1, 2, 5, 8, 10, 12, 15, 20, 25, 50, 75, 100],
    'penalty': ['l1', 'l2']
}

grid_searchlr = GridSearchCV(lr, param_grid, cv=5, scoring='accuracy')
grid_searchlr.fit(x_train, y_train)
```

```python
best_params = grid_searchlr.best_params_
best_model = grid_searchlr.best_estimator_
```

```python
from sklearn.metrics import classification_report
lr= best_model.fit(x_train, y_train)
y_pred = lr.predict(x_test)
report = classification_report(y_test, y_pred)
print(report)
```

```python
accuracy = best_model.score(x_test, y_test)
accuracy
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(x, y, test_size=0.25, random_state=42)

clf = DecisionTreeClassifier()

param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 21],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_searchdt = GridSearchCV(clf, param_grid, cv=5)

grid_searchdt.fit(X_train1, y_train1)

print("Best Parameters: ", grid_searchdt.best_params_)
print("Best Accuracy: ", grid_searchdt.best_score_)
```

```python
clf = DecisionTreeClassifier( criterion = 'gini', max_depth =  5, min_samples_leaf= 2, min_samples_split= 10)

clf.fit(X_train1, y_train1)

y_pred1 = clf.predict(X_test1)

clf.feature_importances_

from sklearn.tree import plot_tree
plt.figure(figsize=(100,175))
plot_tree(clf)
plt.show()
```

```python
print(classification_report(y_test1,y_pred1))
```

SVM

```python
from sklearn.svm import SVC

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf']
}

#SVM classifier
svm_classifier = SVC()

grid_searchsvm = GridSearchCV(svm_classifier, param_grid, cv=5)

grid_searchsvm.fit(X_train1, y_train1)

best_model = grid_searchsvm.best_estimator_

y_predd = best_model.predict(X_test1)

# best hyperparameters
best_params = grid_searchsvm.best_params_
print(f'Best Hyperparameters: {best_params}')

report = classification_report(y_test1, y_predd)
print(report)

```

KNN


```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

param_grid = {'n_neighbors': range(1, 21)}

grid_searchk = GridSearchCV(knn, param_grid, cv=5)
grid_searchk.fit(X_train1, y_train1)

best_k = grid_searchk.best_params_['n_neighbors']
print("Best k:", best_k)
```

```python
knn = KNeighborsClassifier(n_neighbors=14)
knn.fit(X_train1, y_train1)
```

```python
from sklearn.metrics import accuracy_score

y_predk = knn.predict(X_test1)
accuracy = accuracy_score(y_test1, y_predk)
print("Accuracy:", accuracy)


print(classification_report(y_test1, y_predk))
```

Voting

```python
from sklearn.ensemble import VotingClassifier


best_knn = grid_searchk.best_estimator_
best_svm = grid_searchsvm.best_estimator_
best_dt = grid_searchdt.best_estimator_
best_lr = grid_searchlr.best_estimator_

voting_classifier = VotingClassifier(
    estimators=[('knn', best_knn), ('svm', best_svm), ('dt', best_dt), ('lr', best_lr)],
    voting='hard'
)


voting_classifier.fit(X_train1, y_train1)


y_pred_voting = voting_classifier.predict(X_test1)


accuracy_voting = accuracy_score(y_test1, y_pred_voting)
print(f'Voting Classifier Accuracy: {accuracy_voting:.2f}')

report_voting = classification_report(y_test1, y_pred_voting)
print(report_voting)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier


rf_classifier = RandomForestClassifier(random_state=1)

param_grid = {
    'n_estimators': [100, 250, 500, 1000, 1500],
    'max_leaf_nodes': [None, 5, 10, 20, 23],
}

grid_search_rf = GridSearchCV(rf_classifier, param_grid, cv=5)

grid_search_rf.fit(X_train1, y_train1)

best_rf = grid_search_rf.best_estimator_

best_rf_accuracy = best_rf.score(X_test1, y_test1)
print(f'Best Random Forest Accuracy: {best_rf_accuracy * 100:.2f}%')

print("Best Parameters: ", grid_search_rf.best_params_)

y_predrf = best_rf.predict(X_test1)
report_rf = classification_report(y_test1, y_predrf)
print(report_rf)

```
